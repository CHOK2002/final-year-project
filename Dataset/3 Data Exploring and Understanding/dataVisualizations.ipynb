Data Visualizations

Relationship between number of news articles published by month across with all years.

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

# load the dataset
file_path = 'News_Articles_Final.csv'
data = pd.read_csv(file_path)

# convert Date published column to datetime
data['Date_published_dt'] = pd.to_datetime(data['Date published'], errors='coerce')
data['Year'] = data['Date_published_dt'].dt.year
data['Month'] = data['Date_published_dt'].dt.month

# aggregate publication count by month and year
publication_by_month = data.groupby(['Year', 'Month']).size().reset_index(name='Counts')
# plot the news publication with all years
plt.figure(figsize=(10, 5))
sns.lineplot(x='Month', y='Counts', hue='Year', data=publication_by_month, marker='o', palette='tab20')

# adjust the legend to include all years
plt.title('News Articles Publication (All Years)')
plt.xlabel('Month')
plt.ylabel('Number of News Articles Published')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(range(1, 13))  
plt.tight_layout()
plt.show()
```

Relationship between articles text length and number of articles

```python
# plotting the graph of arctile text length distribution 
data['Article_Text_Length'] = data['Article text'].dropna().apply(len)
sns.histplot(data['Article_Text_Length'], bins=30, color="blue")
plt.title('Article Text Length Distribution')
plt.xlabel('Length of Article Text')
plt.ylabel('Numbers of News Articles')
plt.show()
```
