Comparison 1 (News 1 and News 3)

```python
import pandas as pd
import datacompy

# load the datasets
news_articles_1 = pd.read_csv("news_articles_1.csv")
news_articles_3 = pd.read_csv("news_articles_3.csv")

# initialize the DataCompare object
compare = datacompy.Compare(
    news_articles_1,
    news_articles_3,
    join_columns='Article text',  
    abs_tol=0,          
    rel_tol=0,          
    df1_name='News_Articles_1',  
    df2_name='News_Articles_3'  
)

# run the comparison
comparison_results = compare.report()

# print the comparison results
print(comparison_results)
```
